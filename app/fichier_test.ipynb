```python
import re
import unidecode


def removeBlackList(df,apikey):
    blackList = myBacklList(apikey)
    df['name_normalized'] = df['Nom'].apply(lambda x: unidecode.unidecode(re.sub(r'[^\w\s_]', '', x.lower()).replace(' ', '').replace('_','')))
    print(df['name_normalized'])
    # Créer une liste contenant les noms normalisés de la blacklist
    blacklist_names = [unidecode.unidecode(re.sub(r'[^\w\s_]', '', item['properties']['name'].lower()).replace(' ', '').replace('_','')) for item in blackList]

    print(blacklist_names)
    # Sélectionner les lignes du dataframe qui ont une valeur dans la colonne "name_normalized" qui est dans la liste de la blacklist
    # df_filtered = df[~df['name_normalized'].isin(blacklist_names)]
    
    # Sélectionner les lignes du DataFrame qui ont une valeur dans la colonne "name_normalized"
    # qui contient une chaîne de caractères de la liste de la blacklist
    df_filtered = df[df['name_normalized'].str.contains('|'.join(blacklist_names))]
    
    # Supprimer la colonne "name_normalized" du dataframe final
    df = df.drop(df_filtered.index)

    return df
    
```

```python
#import un csv avec pandas et le convertir en dataframe
import pandas as pd
df = pd.read_csv("Black-list-partenaires-OPS-.csv",delimiter="\t")
df = pd.DataFrame(df)
cols = df.columns.values[0].replace(" / ","/").replace("SECTEURS BLACKLISTES","SECTEURS-BLACKLISTES").split(' ')
realCols = []
for e in cols:
    if(e != ''):
        realCols.append(e)
realCols = ['BANQUE/ASSURANCE/FINANCES',
 'DISTRIBUTION',
 'TOURISME',
 'IMMOBILIER',
 'SANTÉ',
 'TRANSPORT',
 'SERVICES',
 'HIGH TECH',
 'SECTEURS-BLACKLISTES']  
```

```python
myDataFrame = pd.DataFrame(columns=realCols)
myDataFrame
```

```python
from itertools import zip_longest

for i in range (0,df.shape[0]):
    values = df[df.columns.values[0]][i].split("  ")
    realvalues = []
    for e in values:
        if(e != ''):
            #condi première ligne ...
            if(i == 0 and e == " Bouygues Telecom FABRICANTS HORLOGERIE"):
                tmp = 'Bouygues Telecom'      
                tmp2 = 'FABRICANTS HORLOGERIE'
                realvalues.append(tmp)
                realvalues.append(tmp2)
            else:    
                realvalues.append(e.strip())
    row = dict(zip_longest(myDataFrame.columns, realvalues, fillvalue=None))
    myDataFrame.loc[i] = row

myDataFrame
```

```python
myDataFrame = myDataFrame.drop(myDataFrame.columns[-1], axis=1)
```

```python
# Créer un nouveau DataFrame vide
new_df = pd.DataFrame(columns=['company_name', 'category', 'blacklist'])

# Parcourir chaque ligne du DataFrame original
for idx, row in myDataFrame.iterrows():
    # Parcourir chaque colonne du DataFrame original
    for col_name in myDataFrame.columns:
        # Si la valeur de la colonne n'est pas nulle et n'est pas dans la colonne SECTEURS-BLACKLISTES
        if not pd.isnull(row[col_name]) and col_name != 'SECTEURS-BLACKLISTES':
            # Ajouter une nouvelle ligne au nouveau DataFrame avec les valeurs correspondantes
            new_row = {'company_name': row[col_name], 'category': col_name, 'blacklist': True}
            new_df = new_df.append(new_row, ignore_index=True)

print(new_df.head())
```

```python
import requests
import json

# Définir l'URL de base pour l'API d'Hubspot
url_base = 'https://api.hubapi.com'

# Définir l'URL pour récupérer les transactions sur le pipeline Sales
url_sales = url_base + '/deals/v1/deal/paged?hapikey=',apikey,'&includeAssociations=true&associations=company'

# Récupérer les transactions sur le pipeline Sales
response_sales = requests.get(url_sales)
deals_sales = json.loads(response_sales.text)['deals']

# Définir l'URL pour récupérer les transactions sur le pipeline Customer
url_customer = url_base + '/crm/v3/objects/deals?hapikey=YOUR_HUBSPOT_API_KEY&properties=name&properties=dealstage&properties=amount&properties=hubspot_owner_id&properties=createdate&properties=dealtype&limit=100'

# Récupérer les transactions sur le pipeline Customer
response_customer = requests.get(url_customer)
deals_customer = json.loads(response_customer.text)['results']

# Comparer les transactions en fonction de leur nom
transaction_names_sales = [deal['properties']['dealname']['value'] for deal in deals_sales]
transaction_names_customer = [deal['properties']['name'] for deal in deals_customer]

# Afficher les transactions manquantes sur le pipeline Customer
missing_transactions = set(transaction_names_sales) - set(transaction_names_customer)
print("Transactions manquantes sur le pipeline Customer:")
print(missing_transactions)
```
